```python
fd["Month"]=df["Order Date"].str[3:5]
fd
```

```python
m={1:"January",2:"Fabruary",3:"March",4:"April",5:"May",6:"June",7:"July",8:"August",9:"September",10:"October",11:"November",12:"December"}
```

```python
fd["Order Date"] = pd.to_datetime(fd["Order Date"], format='%d/%m/%Y')
```

```python
fd["Ship Date"] = pd.to_datetime(fd["Ship Date"], format='%d/%m/%Y')
```

```python
fd["Year"]=fd["Order Date"].dt.year
```

```python
fd["Year"].value_counts()
```

```python
fd["Month"].value_counts()
```

```python
monthly_sales = fd.groupby("Month")["Sales"].sum().reset_index()
monthly_sales.columns = ["Month", "Sales"]
monthly_sales
```

```python
plt.bar(monthly_sales["Month"],monthly_sales["Sales"])
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.xticks(fontsize=10,rotation=90)
plt.yticks(fontsize=10)
plt.show()
```

```python
yearly_sales = fd.groupby("Year")["Sales"].sum().reset_index()
yearly_sales.columns = ["Year", "Sales"]
yearly_sales
```

```python
plt.bar(yearly_sales["Year"],yearly_sales["Sales"])
plt.xlabel("Year")
plt.ylabel("Sales Amount")
plt.xticks(fontsize=10,rotation=90)
plt.yticks(fontsize=10)
plt.show()
```

```python
customer = fd.groupby("Customer Name")["Sales"].sum().reset_index()
customer.columns = ["Customer Name", "Sales"]
customer
```

```python
customer_sorted = customer.sort_values(by="Sales", ascending=False)
customer_sorted
```

```python
customer=customer_sorted.head(20)
```

```python
plt.bar(customer["Customer Name"],customer["Sales"])
plt.xlabel("Customer Name")
plt.ylabel("Sales Amount")
plt.xticks(fontsize=10,rotation=90)
plt.yticks(fontsize=10)
plt.show()
```

```python
category = fd.groupby("Category")["Sales"].sum().reset_index()
category.columns = ["Category", "Sales"]
category
```

```python
plt.bar(category["Category"],category["Sales"])
plt.xlabel("Category Name")
plt.ylabel("Sales Amount")
plt.xticks(fontsize=10,rotation=90)
plt.yticks(fontsize=10)
plt.show()
```

```python
plt.bar(category["Category"], category["Sales"], color='blue', alpha=0.7)
plt.xlabel("Category Name")
plt.ylabel("Sales Amount")
plt.title("Sales Amount by Category")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.pie(category["Sales"], labels=category["Category"], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # To view the circle as a circle
plt.title("Sales Distribution by Category")
plt.show()
```
